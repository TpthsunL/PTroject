- 1. 일자별 시세를 페이지별로 수집

```python
def get_day_list(item_no, page_no):
    """
    일자별 시세를 페이지별로 수집
    """ 
    url=f'https://finance.naver.com/item/board.nhn?code={item_no}&page={page_no}'
    headers= {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36 Edg/86.0.622.51'}
    response = requests.get(url, headers = headers)
    html = bs(response.text, 'lxml')
    temp = html.select('table')
    df_day = pd.read_html(str(temp))
    df_day = df_day[1].dropna(how='all', axis=1)
    df_day = df_day.dropna()
    return df_day
#f : f-string의 줄임말로, {}를 통해서 변수를 가독성이 좋도록 표현할 수 있게끔 파이썬에 만든 기능이다. 
```

```python
item_no=950130
page_no=3
get_day_list(950130,4)
```

```python
import time
from tqdm import trange

page_no=1
start_page=1
end_page=10

item_list=[]
for page_no in trange(start_page, end_page+1):
    df_temp_day =get_day_list(item_no, page_no)
    item_list.append(df_temp_day)
```
